### –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –≤–∞—à–æ–≥–æ –Ω–∞–±–æ—Ä—É –¥–∞–Ω–∏—Ö

```python
df=pd.read_excel("–ü–æ–≤—ñ—Å—Ç–∫–∏_—Å–∏—Ä—ñ –¥–∞–Ω—ñ.xlsx")
df.head()
```

### –ó–∞–≥–∞–ª—å–Ω—ñ —Ç—Ä–∞–Ω—Å–æ—Ñ—Ä–º–∞—Ü—ñ—ó —Ä–µ–π—Ç–∏–Ω–≥—É —ñ –Ω—ñ–∫–Ω–µ–π–º—ñ–≤

#### –ó–∞–±–∏—Ä–∞—î–º–æ —Ü–∏—Ñ—Ä–∏ –ø–µ—Ä–µ–¥ –Ω–∞–∑–≤–æ—é –∫–∞–Ω–∞–ª—É –≤ –Ω–æ–≤—É –∫–æ–ª–æ–Ω–∫—É

```python
df['rank'] = df['channel search'].str.extract(r'(\d+)', expand=False)
df
```

#### –ó–∞–±–∏—Ä–∞—î–º–æ —Ü–∏—Ñ—Ä–∏ —Å–ø–µ—Ä–µ–¥—É –Ω–∞–∑–≤–∏ –∫–∞–Ω–∞–ª—É

```python
df['channel search'] = df['channel search'].str.replace(r'^\d+?', '').astype(str)
df['channel search'] = df['channel search'].str.replace(r'^\d+?', '').astype(str) #—â–µ —Ä–∞–∑ –Ω–∞ –≤—Å—è–∫–∏–π –≤–∏–ø–∞–¥–æ–∫
df
```

#### –í–∏—Ç—è–≥—É—î–º–æ –Ω—ñ–∫ –∫–∞–Ω–∞–ª—É —ñ –≤–∏–∫–∏–¥–∞—î–º–æ –¥—É–±–ª—ñ–∫–∞—Ç–∏

```python
df['channel_nickname'] = df['channel search'].str.split(r'@').str.get(1)
df['channel search'] = df['channel search'].str.split(r'@').str.get(0)
df['channel_nickname'] = df['channel_nickname'].str.replace(r'\$\d+?', '').astype(str) #–≤–±–∏–≤–∞—î–º–æ –∑–Ω–∞—á–æ–∫ –¥–æ–ª–∞—Ä–∞

df.drop_duplicates(subset=['channel_nickname'], keep='last')

df
```

### –ù–æ—Ä–º–∞–ª—ñ–∑–∞—Ü—ñ—è –ø—ñ–¥–ø–∏—Å–Ω–∏–∫—ñ–≤

#### –ù–æ—Ä–º–∞–ª—ñ–∑—É—î–º–æ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø—ñ–¥–ø–∏—Å–Ω–∏–∫—ñ–≤ —è–∫ —á–∏—Å–ª–æ

```python
#—Ç–∏—Å—è—á—ñ —á–∏—Ç–∞—á—ñ–≤
k_subscribers=df[df.Subscribers.str.contains('k', regex= True, na=False)]
k_subscribers["Subscribers"]=k_subscribers["Subscribers"].str.replace(r'.', '').astype(str) #–∑–∞–±–∏—Ä–∞—î–º –∫—Ä–∞–ø–∫—É
k_subscribers["Subscribers"]=k_subscribers["Subscribers"].str.replace(r'k', '00').astype(str) #–¥–æ–¥–∞—î–º –¥–≤–∞ –Ω—É–ª—è –∑–∞–º—ñ—Å—Ç—å k

#–º—ñ–ª—å–π–æ–Ω–∏ —á–∏—Ç–∞—á—ñ–≤
m_subscribers=df[df.Subscribers.str.contains('m', regex= True, na=False)]
m_subscribers["Subscribers"]=m_subscribers["Subscribers"].str.replace(r'.', '').astype(str) #–∑–∞–±–∏—Ä–∞—î–º –∫—Ä–∞–ø–∫—É
m_subscribers["Subscribers"]=m_subscribers["Subscribers"].str.replace(r'm', '00000').astype(str) #–¥–æ–¥–∞—î–º –¥–≤–∞ –Ω—É–ª—è –∑–∞–º—ñ—Å—Ç—å k

#–º—ñ–ª—å—è—Ä–¥–∏ —á–∏—Ç–∞—á—ñ–≤
b_subscribers=df[df.Subscribers.str.contains('b', regex= True, na=False)]
b_subscribers["Subscribers"]=b_subscribers["Subscribers"].str.replace(r'.', '').astype(str) #–∑–∞–±–∏—Ä–∞—î–º –∫—Ä–∞–ø–∫—É
b_subscribers["Subscribers"]=b_subscribers["Subscribers"].str.replace(r'k', '00000000').astype(str) #–¥–æ–¥–∞—î–º –¥–≤–∞ –Ω—É–ª—è –∑–∞–º—ñ—Å—Ç—å k

#–ø—Ä–æ—Å—Ç—ñ —Ü–∏—Ñ—Ä–∏
less1000_subcsribers=df[(~df["Subscribers"].str.contains('b', regex= True, na=False)) & (~df["Subscribers"].str.contains('m', regex= True, na=False)) & (~df["Subscribers"].str.contains('k', regex= True, na=False))]

#–∑–≤–æ–¥–∏–º–æ –≤ –æ–¥–∏–Ω –Ω–∞–±—ñ—Ä –¥–∞–Ω–∏—Ö
df = pd.concat([k_subscribers, m_subscribers, b_subscribers, less1000_subcsribers])
df['Subscribers'].astype('int64')

df
```

### –ù–æ—Ä–º–∞–ª—ñ–∑—É—î–º–æ —Å–µ–º–∏–¥–µ–Ω–Ω–∏–π –ø—Ä–∏—Ä—ñ—Å—Ç –ø—ñ–¥–ø–∏—Å–Ω–∏–∫—ñ–≤
